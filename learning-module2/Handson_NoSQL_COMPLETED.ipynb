## One Advantage of NoSQL: Nested Documents

NoSQL databases such as MongoDB allow storing of nested documents. Let's examine the advertising focus of Daniel Brian Advertising. So, every information for one entity can be effectively stored together, so there is no need for complex schemas!

```python
dba_focus = db.Advertising_Agencies.find_one({"Name": "Daniel Brian Advertising"})["Advertising_Focus"]
dba_focus
```

Five focus items are stored. Compare this to the focus items of Anchour.

```python
anchour_focus = db.Advertising_Agencies.find_one({"Name": "Anchour"})["Advertising_Focus"]
anchour_focus
```

A NoSQL database allows storing only relevant information. Imagine having an extensive list of agencies with various focus areas. A SQL table would consist of many columns, one for each focus. Most of the columns would be filled with mostly zeros for each row.

The attributes for each document are optional. Check the following query regarding reviews

```python
anchour_reviews = db.Advertising_Agencies.find_one({"Name": "Anchour"})["Reviews"]
anchour_reviews
```

## One Disadvantage of NoSQL: Specific Syntax

But NoSQL databases have disadvantages too. Each NoSQL has a specific Syntax and as opposed to SQL that serves as "industry standard" chosing a specific NoSQL database may create complex dependencies.

For example, lets create a the equivalent of a "WHERE" statement and query all agencies founded after the year 1990.

```python
all_agencies_after_1990 = db.Advertising_Agencies.find({"Founded": {"$gte": 1990}})
all_agencies_after_1990
```

**"find({"Founded": {"$gte": 1990}})"** is the specific command for MongoDB for the SQL statement "WHERE Founded >= 1990". In other NoSQL document databases the syntax might be completely different

Also, the results refer to a cursor object, i.e., "pointers" to the specific instances in the database (instead of retrieving the results directly as in the SQL Handson)

For getting the results, we need to iterate over the cursor.

```python
for agency in all_agencies_after_1990:
  print(agency)
```

```python
all_agencies_after_1990 = db.Advertising_Agencies.find({"Founded": {"$gte": 1990}})
for agency in all_agencies_after_1990:
  print(agency["Name"],"has been founded in", agency["Founded"])
```
